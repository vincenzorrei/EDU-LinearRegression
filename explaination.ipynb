```python
import numpy as np
import matplotlib.pyplot as plt
```

# Il caso Instagram
Supponiamo di avere osservato di aver fatto diverse pubblicità sponsorizzate su instagram e di aver tenuto traccia delle visualizzazioni ottenute.

In particolare, considereremo X come il budget per la sponsorizzata e y come il numero di view per quella sponsorizzata.

| x : budget   | y : views    |
|--------------|--------------|
|      10      |      80      |
|      70      |     260      |
|      90      |     320      |
|      20      |     110      |
|      35      |     155      |

Qual è la relazione che lega le il budget alle views?


```python
x  =[10, 70, 90, 20, 35]
y = [80, 260, 320, 110, 155]
```

```python
def regress_prediction(x):
    """Questa funzione restituisce la predizione del modello di regressione lineare"""
    B_0 = 50
    B_1 = 3
    y_hat = B_0 + B_1 * x
    return y_hat
```

```python
for i in range(len(x)):
    print(f"Per x = {x[i]} la predizione del modello è {regress_prediction(x[i])}")
```

```python
# Calcolo dei coefficienti della linea usando numpy.polyfit
coefficients = np.polyfit(x, y, 1)

# Funzione per la linea
line_function = np.poly1d(coefficients)

# Creazione di nuovi punti per tracciare la linea
x_line = np.linspace(min(x)-50, max(x)+50, 100)
y_line = line_function(x_line)

# Grafico dei punti e della linea
plt.plot(x_line, y_line, color='C1', label='Linea di adattamento')
plt.scatter(x, y, label='Punti dati')


plt.xlabel('X')
plt.ylabel('Y')
plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)
plt.legend()
plt.grid()
plt.title('Linea creata dai punti dati')
plt.show()
```
